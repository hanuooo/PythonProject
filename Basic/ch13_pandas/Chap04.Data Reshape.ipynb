```python
dataIn = './../dataIn/'
dataOut = './../dataOut/'
myencoding = 'CP949'
```

```python
filename = dataIn + '가전제품판매량01.csv'
homeware01 = pd.read_csv(filename, encoding=myencoding, index_col='지역')
homeware01
```

```python
filename = dataIn + '가전제품판매량02.csv'
homeware02 = pd.read_csv(filename, encoding=myencoding, index_col='지역')
homeware02
```

```python
columnInfo01 = homeware01.columns # 상반기 컬럼 정보
columnInfo02 = homeware02.columns # 하반기 컬럼 정보
```

```python
print('양쪽에 공통적으로 들어 있는 제품')
columnInfo01.intersection(columnInfo02) # pd.Index.intersection() : 공통 인덱스 추출(교집합)
```

```python
print('상반기에만 판매된 제품')
print(columnInfo01.difference(columnInfo02)) # pd.Index.difference()   : 한쪽에만 있는 인덱스 추출(차집합)

print('하반기에만 판매된 제품')
print(columnInfo02.difference(columnInfo01))
```

```python
print(columnInfo01.union(columnInfo02)) # pd.Index.union() : 인덱스 합집합
```

```python
print('분기별 판매량 결합')
result = pd.concat([homeware01, homeware02], join='inner')
result
```

```python
print('양쪽에 공통적이지 않은 한목은 NaN 데이터가 됩니다.')
result = pd.concat([homeware01, homeware02], axis=0) # outer join
result
```

```python
print('열방향으로 합치기.')
result = pd.concat([homeware01, homeware02], axis=1)
result
```

```python
filename = dataIn + 'movie_show.csv'
movieshow = pd.read_csv(filename, encoding='CP949')
movieshow
```

```python
movieshow_melt = pd.melt(movieshow, id_vars=['gu', 'year', 'month'], var_name='genre', value_name='qty')
movieshow_melt
```

```python
# aggfunc 매개 변수(집계 함수)의 기본 값은 평균(mean) 입니다.
movieshow_pivot = movieshow_melt.pivot_table(index=['gu', 'year'], columns='month', values='qty',aggfunc='sum')
movieshow_pivot
```

```python
# 장르는 갯수, 수량(qty)는 평균 정보
dict_aggfunc = {'genre':'count', 'qty':'mean'}

movieshow_pivot = movieshow_melt.pivot_table(
    index=['gu', 'year'],
    columns='month',
    values=['genre', 'qty'],
    aggfunc=dict_aggfunc)
movieshow_pivot
```

```python
# 특정 칼럼에 대하여 여러 개의 함수 지정이 가능합니다.
dict_aggfunc = {'qty':['mean', 'sum']}

movieshow_pivot = movieshow_melt.pivot_table(
    index=['gu', 'year'],
    columns='month',
    values='qty',
    aggfunc=dict_aggfunc)
movieshow_pivot
```

```python
# 내부와 외부 색인을 합쳐서 간단하게 만들기
outer = movieshow_pivot.columns.get_level_values(0) # 외부 색인
inner = movieshow_pivot.columns.get_level_values(1) # 내부 색인

hangul_dict = {'mean':'평균', 'sum':'총합', 'max':'최대'}
newOuter = [hangul_dict[asdf] for asdf in outer]

inner = inner.astype(str)

newColumns = inner + '_' + newOuter
movieshow_pivot.columns = newColumns
movieshow_pivot
```
