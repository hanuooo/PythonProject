```python
store , price = '메가커피' , 100
print(f'{store} 단가가 {price} 이하인 품목들')
result02 = myframe02.loc[myframe02[store] <= price]
result02
# myframe02[store] <= su
```

```python
store , price = '이디야커피' , 180
print(f'{store} 단가가 {price} 이하인 품목들')
result02 = myframe02.loc[myframe02[store] == price]
result02
```

```python
condition01 = (myframe02['스타벅스'] >= 70)
condition01
```

```python
condition02 = (myframe02['메가커피'] >=140)
condition02
```

```python
# and, or, not 연산자는 1대1 비교라서 사용하지 못합니다.
# 대신 element wise(원소별 비교) 를 위하여 &(and), |(or), 앞에 ~(not)을 사용합니다.
boolResult = condition01 & condition02
result02 = myframe02.loc[boolResult]
result02
```

```python
print('양쪽 결과 통합')
condition = pd.DataFrame([condition01, condition02])
condition.T
```

```python
print('모든 항목들이 True인거 찾기')
condition.all()
```

```python
print('항목들 중에서 하나라도 True 인거 찾기')
condition.any()
```

```python
print('all() 함수를 사용하여 모든 조건이 True인거 추출하기')
result02 = myframe02.loc[condition.all()]
result02
```

```python
print('any() 함수를 사용하여 조건이 하나라도 True인거 추출하기')
result02 = myframe02.loc[condition.any()]
result02
```
