<h6><font color="BlueViolet">시리즈의 값 읽어 오기</font></h6>

```python
myindex02 = ['아메리카노', '에스프레소', '카페라떼', '카페모카', '콜드브루', '바닐라라떼', '마끼아또']
mylist02 = [50, 30, 20, 60, 40, 80, 70]
myserise02 = pd.Series(data=mylist02, index=myindex02)
myserise02.to_frame()
```

```python
myserise02['카페모카'] # scalar로 반환
```

```python
myserise02[['카페모카']] # Series로 반환
```

```python
myserise02['카페모카':'에스프레소']
```

```python
# 콜론 기호는 연속된 데이터 추출
myserise02['에스프레소':'카페모카'] # 문자열 라벨링 : from <= 범위 <= to
```

```python
myserise02[['마끼아또', '카페모카']] # 콤마 기호는 서로 떨어진 데이터 추출
```

```python
myserise02.iloc[3] # i (index) loc (location) 정수 위치 기반 접근 방식
```

```python
myserise02.iloc[0:7:3] # 슬라이싱
```

```python
myserise02.iloc[3:6]
```

<h6><font color="BlueViolet">시리즈에 값 쓰기</font></h6>

```python
myserise02.iloc[2] = 2000
myserise02.iloc[2:5] = 5000
myserise02[['아메리카노', '카페모카']] = 3000
print(myserise02)
```

```python
myserise02.iloc[0::2] = 7000
print(myserise02)
```

```python
result = myserise02[myserise02 <= 4000]
print(result)
```

<h3><font color="red">시리즈 연산</font></h3>

<h6><font color="BlueViolet">시리즈 샘플 만들기</font></h6>

```python
myindex31= ['에스프레소', '아메리카노', '마키아또', '바닐라라떼']
mylist31 = [30, 40, 50 ,60]

myindex32 = ['에스프레소', '아메리카노', '마키아또', '콜드브루']
mylist32 = [20, 40, 60, 70]

myseries31 = pd.Series(data=mylist31, index=myindex31)
myseries32 = pd.Series(data=mylist32, index=myindex32)
```

```python
myseries31
```

```python
myseries32
```

```python
newseries = myseries31 + myseries32
# NaN : Not a Number ( 숫자가 아니다 , 결척지(빠진 데이터))
print(newseries)
```

```python
newseries = myseries31.sub(myseries32, fill_value=0)
print(newseries)
```

```python
newseries = myseries31.mul(myseries32, fill_value=0)
print(newseries)
```

<h3><font color="red">유일값과 빈도 파악, 색인 필터링</font></h3>

<h6><font color="BlueViolet">unique 메소드 사용</font></h6>

```python
mylist04 = ['카페모카', '에스프레소', '에스프레소', '카페라떼', '에스프레소', '에스프레소', '아메리카노', '에스프레소', '카페모카', '카페모카']
myserise04 = pd.Series(mylist04)
myserise04 = np.array(myserise04)

type(myserise04)
print(myserise04.head())
print(myserise04.unique())
print(myserise04.value_counts())
```

```python
myunique04 = myserise04.unique()
myunique04
```

```python
myserise04.value_counts() # 빈도수 파악해서 역순으로 정렬
```

```python
myserise04.value_counts(normalize=True) # 빈도수를 비율로 보여주기
```

```python
mask04 = myserise04.isin(['카페모카', '카페라떼'])
mask04
```

```python
myserise04[mask04]
```

<h3><font color="red">그래프 그리기</font></h3>

```python
myindex05 = ['아메리카노', '에스프레소', '카페라떼', '카페모카', '콜드부르', '바닐라라떼', '마끼아또']
mylist05 = [5000, 3000 , 2000, 6000, 4000, 8000, 7000]
myseries05 = pd.Series(data=mylist05, index=myindex05)
myseries05
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
myseries05.plot(kind='line', grid=True, rot=20)
plt.xlabel('상품명', fontsize = 14)
plt.ylabel('단가', fontsize = 14)
plt.title('품목별 단가 꺾은 선', fontsize = 18)
```

```python
myseries05.plot(kind='bar', grid=True, rot=0, color=['r','g','b','y','m'])
plt.xlabel('상품명', fontsize = 14)
plt.ylabel('단가', fontsize = 14)
plt.title('품목별 단가 꺾은 선', fontsize = 18)
```

```python
myseries05.plot(kind='barh', grid=True, rot=0, color=['r','g','b','y','m'])
plt.xlabel('상품명', fontsize = 14)
plt.ylabel('단가', fontsize = 14)
plt.title('품목별 단가 꺾은 선', fontsize = 18)
```

```python
myseries05.plot(kind='pie', grid=True, rot=0, color=['r', 'g', 'b', 'y', 'm'])
plt.title('품목별 단가 꺾은 선', fontsize=18)
```
